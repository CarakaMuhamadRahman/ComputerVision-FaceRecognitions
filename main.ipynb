```python
import cv2
import pickle as pkl
import numpy as np
import os
import csv
import time
from datetime import datetime
```

<h1>Take Video</h1>

```python
video = cv2.VideoCapture(0)
feature_extractions = cv2.CascadeClassifier('./Data/haarcascade_frontalface_default.xml')

faces_data = []
i = 0
name = input('Enter your name: ')

while True:
    ret, frame = video.read()
    grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    face = feature_extractions.detectMultiScale(grey, 1.3, 5)
    
    for x,y,w,h in face:
        crop_image = frame[y:y+h, x:x+w, :]
        resized_image = cv2.resize(crop_image, (50,50))
        print(resized_image)
        if len(faces_data) <= 100 and i%10==0:
            faces_data.append(resized_image)
        i += 1
        
        cv2.putText(frame, str(len(faces_data)), (50,50), cv2.FONT_HERSHEY_COMPLEX, 2,
                    (50,50,255), 2)
        
        cv2.rectangle(frame, (x,y), (x+w, y+h), (50, 50, 255), 2)
    cv2.imshow("Camera", frame)

    k = cv2.waitKey(1)

    if k == ord('q') or len(faces_data)==100:
        break

video.release()
cv2.destroyAllWindows()
```

<h1>Training Using K-NN</h1>

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
with open('Data/names.pkl', 'rb') as f:
    y_train = pkl.load(f)

with open('Data/faces_data.pkl', 'rb') as f:
    X_train = pkl.load(f)
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3)

model_knn.fit(X_train, y_train)
```

<h1>Testing</h1>

```python
from win32com.client import Dispatch

def speak(speaks):
    speak = Dispatch(("SAPI.SpVoice"))
    speak.Speak(speaks)
```

```python
video = cv2.VideoCapture(0)
feature_extractions = cv2.CascadeClassifier('./Data/haarcascade_frontalface_default.xml')

col_names = ['NAME', 'DATE', 'TIME']

while True:
    ret, frame = video.read()
    grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    face = feature_extractions.detectMultiScale(grey, 1.3, 5)
    
    for x,y,w,h in face:
        crop_image = frame[y:y+h, x:x+w, :]
        resized_image = cv2.resize(crop_image, (50,50)).flatten().reshape(1, -1)
        output = model_knn.predict(resized_image)
        ts = time.time()
        dateTime = datetime.fromtimestamp(ts).strftime("%d-%m-%Y")
        timeStamp = datetime.fromtimestamp(ts).strftime("%H:%M:%S")
        exist = os.path.isfile('Attendance/Attendance_' + dateTime + '.csv')

        cv2.putText(frame, f'{str(output[0])}: {model_knn.predict_proba(resized_image)}', (x,y-15), cv2.FONT_HERSHEY_COMPLEX, 1, (255,255,255), 2)
        
        cv2.rectangle(frame, (x,y), (x+w, y+h), (50, 50, 255), 2)

        attendance = [str(output[0]), str(dateTime), str(timeStamp)]
    cv2.imshow("Frame", frame)

    k = cv2.waitKey(1)
    if k == ord('o'):
        speak('Attendance Taken')
        time.sleep(5)
        if exist:
            with open('Attendance/Attendance_' + dateTime + '.csv', '+a') as f:
                writer = csv.writer(f)
                writer.writerow(attendance)
            f.close()
        else:
            with open('Attendance/Attendance_' + dateTime + '.csv', '+a') as f:
                writer = csv.writer(f)
                writer.writerow(col_names)
                writer.writerow(attendance)
            f.close()
    if k == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
